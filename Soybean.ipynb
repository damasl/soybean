## Обрабатываем пропущенные значения

Если данные содержат пропущенные значения, то имеется две простые альтернативы:
- удалить столбцы с такими значениями (`data = data.dropna(axis = 1)`),
- удалить строки с такими значениями (`data = data.dropna(axis = 0)`).

После этого данных может стать совсем мало.

Куда более эффективным (и в то же время по-прежнему простым) методом обычно является заполнение пропущенных значений для каждой количественной переменной их средним или медианным значением. 
Для категориальных признаков пропущенные значения можно заполнить самым популярным значением в столбце.

```python
data_describe = data.describe(include = [object])
for c in categorical_columns:
    data[c] = data[c].fillna(data_describe[c]['top'])
```

```python
data.describe(include = [object])
```

```python
for c in categorical_columns:
    print(c, data[c].unique())
```

## Векторизация

```python
binary_columns    = [c for c in categorical_columns if data_describe[c]['unique'] == 2]
nonbinary_columns = [c for c in categorical_columns if data_describe[c]['unique'] > 2 and c!='class']
print(binary_columns, nonbinary_columns)
```

```python
data = data.replace(to_replace = ["0","1"], value=[0,1])
```

```python
for b in categorical_columns:
    print(b, data[b].unique())
```

К небинарными признакам применим метод векторизации, который заключается в следующем.

Признак j, принимающий s значений, заменим на s признаков, принимащих значения 0 или 1, в зависимости от того, чему равно значение исходного признака j.

Например, в нашей задаче признак date принимает 7 различных значений:

```python
data['date'].unique()
```

Заменим признак date 7 признаками: 0,1,2,3,4,5,6 (april,may,june,july,august,september,october).

Если признак date принимает значение 0, то признак date_0 равен 1, а все остальные равны 0.
Если признак date принимает значение 1, то признак date_1 равен 1, а все остальные равны 0.
и т.д.

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

```python
data['class'].unique()
```

## Соединяем всё в одну таблицу

```python
data = pd.concat((data_nonbinary, data[binary_columns], data['class']), axis = 1)
print(data.shape)
```

```python
data.describe()
```

## X и y

```python
X = data.drop(('class'), axis = 1) # выбрасываем столбец 'class'
y = data['class']
feature_names = X.columns
```

```python
feature_names
```

```python
print(X.shape)
print(y.shape)
N, d = X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

## $k$NN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
err_test = 1 - knn.score(X_test, y_test)
```

```python
err_test
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, y_test_predict)
print(confusion_matrix(y_test, y_test_predict))
```

Метод не мог определить один из классов, поражение гербицидом, т.к. представители данного класса не попали в обучающую выборку, а, значит, и не могли оказаться среди ближайших соседей. 

```python
results = {};
for i in range(8, 1, -2):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    y_train_predict = knn.predict(X_train)
    y_test_predict = knn.predict(X_test)
    err_train = np.mean(y_train != y_train_predict)
    err_test  = np.mean(y_test  != y_test_predict)
    results[i] = (err_train, err_test)
print(results)
```

Как можно заметить, на данных по сое метод k-ближайших соседей работает не очень хорошо, и результат его работы сильно зависит от выбора параметра k.
